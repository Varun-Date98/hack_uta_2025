```python
import os
import json
import pandas as pd
import google.generativeai as genai
```

```python
downloads_path = r"C:/Users/varun/Downloads/New Folder"
files = [f for f in os.listdir(downloads_path) if f.endswith(".csv")]
```

```python
data_frames = []

for file in files:
    data_frames.append(pd.read_csv(os.path.join(downloads_path, file)))

df = pd.concat(data_frames, ignore_index=True)
```

```python
df.head()
```

```python
df.to_csv("./news_data.csv")
```

```python
genai.configure(api_key="...")  # Add your key here
model = genai.GenerativeModel("gemini-pro")

def get_topics(content):
    res = []
    for text in content["content"]:
        response = model.generate_content(f"Return a list with 5 one word general topics like (globe, sports, polytics, hollywood etc. No country names) for the following article \n{text}")
        
        if response and hasattr(response, "text") and response.text:
            res.append(", ".join([text[2:] for text in response.text.split("\n")]))
        else:
            res.append("No Topic")
    
    return res
```

```python
test_df = df.iloc[:500].copy()
```

```python
test_df["topics"] = [""] * len(test_df.index)
```

```python
test_df.head(10)
```

```python
import time
```

```python
for i in range(0, len(test_df.index), 15):
    batch = test_df.iloc[i:i+15]
    test_df.loc[i:i+14, "topics"] = get_topics(batch)
    print(f"Processed {i + 15} records")
    time.sleep(65)
```
